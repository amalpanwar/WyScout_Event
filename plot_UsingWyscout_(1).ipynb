```python
# Functions to extract start and end coordinates with error handling
def extract_start_x(pos):
    return pos[0]['x'] if len(pos) > 0 else None

def extract_start_y(pos):
    return pos[0]['y'] if len(pos) > 0 else None

def extract_end_x(pos):
    return pos[1]['x'] if len(pos) > 1 else None

def extract_end_y(pos):
    return pos[1]['y'] if len(pos) > 1 else None

# Apply functions to extract coordinates
df_events['x_start'] = df_events['positions'].apply(extract_start_x)
df_events['y_start'] = df_events['positions'].apply(extract_start_y)
df_events['x_end'] = df_events['positions'].apply(extract_end_x)
df_events['y_end'] = df_events['positions'].apply(extract_end_y)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('dark_background')
pitch = Pitch(line_color = "white", pitch_type='wyscout')
fig, ax = pitch.draw(figsize=(10, 7))
# pitchLengthX = 100
# pitchWidthY = 100

for i,thepass in passes.iterrows():
    #if pass made by Lucy Bronze
    if thepass['playerId']==397174:
        x=thepass['x_start']
        y=thepass['y_start']
        x_end = thepass['x_end']
        y_end = thepass['y_end']
        #plot circle
        passCircle=plt.Circle((x,y),2,color="white")
        passCircle.set_alpha(.2)
        ax.add_patch(passCircle)
        dx=thepass['x_end']-x
        dy=thepass['y_end']-y
        #plot arrow
        passArrow=plt.Arrow(x,y,dx,dy,width=2,color="white")
        ax.add_patch(passArrow)

ax.set_title("Scott McTominay Passes agaisnt Watford (13-05-2018)", fontsize = 24)
fig.set_size_inches(10, 7)
plt.show()
```

```python
plt.style.use('dark_background')
pitch = Pitch(line_color = "white", pitch_type='wyscout')
fig, ax = pitch.draw(figsize=(10, 7))
#Size of the pitch in yards (!!!)
pitchLengthX = 100
pitchWidthY = 100
#Plot the shots by looping through them.
for i,shot in shots.iterrows():
    #get the information
    x=shot['x_start']
    y=shot['y_start']
    #goal=shot['outcome_name']=='Goal'
    teamId=shot['teamId']
    #set circlesize
    circleSize=2
    #plot ManUnited
    if (teamId==team1):
        #if goal:
            shotCircle=plt.Circle((x,y),circleSize,color="red")
            plt.text(x+1,y-2,shot['shortName'])
        # else:
        #     shotCircle=plt.Circle((x,y),circleSize,color="red")
        #     shotCircle.set_alpha(.2)
    #plot Watford
    else:
        #if goal:
            shotCircle=plt.Circle((pitchLengthX-x,pitchWidthY-y),circleSize,color="green")
            plt.text(pitchLengthX-x+1,pitchWidthY-y - 2 ,shot['shortName'])
        # else:
        #     shotCircle=plt.Circle((pitchLengthX-x,pitchWidthY - y),circleSize,color="blue")
        #     shotCircle.set_alpha(.2)
    ax.add_patch(shotCircle)
#set title
fig.suptitle("ManUnited (red) and Watford (Green) shots", fontsize = 24)
fig.set_size_inches(10, 7)
plt.show()
```
